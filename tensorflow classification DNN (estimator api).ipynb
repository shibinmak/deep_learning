```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
data = pd.read_csv('pima-indians-diabetes.csv')
raw_data = pd.read_csv('pima-indians-diabetes.csv')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
lis= ['Number_pregnant', 'Glucose_concentration', 'Blood_pressure', 'Triceps',
       'Insulin', 'BMI', 'Pedigree']
```

```python
data[lis]= scaler.fit_transform(data[lis])
```

```python
x= data.drop('Class',axis=1)
```

```python
y = data['Class']
```

```python
 for i in (lis):
     a=  "{}".format(i)
     b= " = tf.feature_column.numeric_column('{}')".format(i)
     print(a+b)
     
     
```

```python
Number_pregnant = tf.feature_column.numeric_column('Number_pregnant')
Glucose_concentration = tf.feature_column.numeric_column('Glucose_concentration')
Blood_pressure = tf.feature_column.numeric_column('Blood_pressure')
Triceps = tf.feature_column.numeric_column('Triceps')
Insulin = tf.feature_column.numeric_column('Insulin')
BMI = tf.feature_column.numeric_column('BMI')
Pedigree = tf.feature_column.numeric_column('Pedigree')
Age = tf.feature_column.numeric_column('Age')
```

```python
Group = tf.feature_column.categorical_column_with_hash_bucket('Group',hash_bucket_size=10)
```

```python
age_bucket = tf.feature_column.bucketized_column(Age,boundaries=[20,30,40,50,60,70])
```

```python
feat_col = lis
```

```python
feat_col.append('age_bucket')
```

```python
feat_col.append('Group')
```

```python
feat_col  = ','.join(feat_col)
```

```python
feat_cols = feat_col.replace("'","")
```

```python
print(feat_cols)
```

```python
feat_cols =[Number_pregnant,Glucose_concentration,Blood_pressure,Triceps,Insulin,BMI,Pedigree,age_bucket,Group]
```

```python
type(feat_cols)
```

```python
print(feat_cols)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
train = model.train(input_fn=input_func,steps=1000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,shuffle=False,num_epochs=1,batch_size=10)
```

```python
train_eval =model.evaluate(input_fn=eval_input_func,steps=1000)
```

```python
train_eval
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=x_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
prediction = model.predict(input_fn=pred_input_func)
```

```python
group_dense=tf.feature_column.embedding_column(Group, dimension=4)
```

```python
feat_colsd= feat_col.replace("'","")
```

```python
feat_colsd = feat_colsd.replace('Group','group_dense')
```

```python
feat_colsd = [Number_pregnant,Glucose_concentration,Blood_pressure,Triceps,Insulin,BMI,Pedigree,age_bucket,group_dense]
```

```python
dnn_model = tf.estimator.DNNClassifier(hidden_units=[10,10,10],feature_columns=feat_colsd,n_classes=2)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
dnn_model.train(input_fn=input_func,steps=10)
```

```python
eval_inputfunc = tf.estimator.inputs.pandas_input_fn(x=x_test,y=y_test,batch_size=10,num_epochs=10,shuffle=False)
```

```python
eval = dnn_model.evaluate(input_fn=eval_inputfunc,steps=10)
```
