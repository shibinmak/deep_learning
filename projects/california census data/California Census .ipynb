```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import tensorflow as tf
```

```python
data = pd.read_csv('census_data.csv')
```

```python
data.head()
```

```python
data['income_bracket'].unique()
```

```python
data['income_bracket'] = data['income_bracket'].apply(lambda label:label==' <=50K')
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x= data.drop('income_bracket',axis=1)
```

```python
y=data['income_bracket']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x.columns
```

```python
num_value=['age','education_num','capital_gain','capital_loss','hours_per_week']
```

```python
cat_value= [x for x in x.columns if x not in num_value]
```

```python
cat_value
```

```python
tf.feature_column.categorical_column_with_hash_bucket('{}',hash_bucket_size=1000)
```

```python
for i in num_value:
    a= '{}'.format(i) + " = tf.feature_column.numeric_column('{}')".format(i)
    print (a)
```

```python
for i in cat_value:
    a= '{}'.format(i) + " = tf.feature_column.categorical_column_with_hash_bucket('{}',hash_bucket_size=1000)".format(i)
    print (a)
```

```python
age = tf.feature_column.numeric_column('age')
education_num = tf.feature_column.numeric_column('education_num')
capital_gain = tf.feature_column.numeric_column('capital_gain')
capital_loss = tf.feature_column.numeric_column('capital_loss')
hours_per_week = tf.feature_column.numeric_column('hours_per_week')
workclass = tf.feature_column.categorical_column_with_hash_bucket('workclass',hash_bucket_size=1000)
education = tf.feature_column.categorical_column_with_hash_bucket('education',hash_bucket_size=1000)
marital_status = tf.feature_column.categorical_column_with_hash_bucket('marital_status',hash_bucket_size=1000)
occupation = tf.feature_column.categorical_column_with_hash_bucket('occupation',hash_bucket_size=1000)
relationship = tf.feature_column.categorical_column_with_hash_bucket('relationship',hash_bucket_size=1000)
race = tf.feature_column.categorical_column_with_hash_bucket('race',hash_bucket_size=1000)
gender = tf.feature_column.categorical_column_with_hash_bucket('gender',hash_bucket_size=2)
native_country = tf.feature_column.categorical_column_with_hash_bucket('native_country',hash_bucket_size=1000)
```

```python
feat= ','.join(x.columns)
```

```python
print(feat)
```

```python
feat_cols=[age,workclass,education,education_num,marital_status,occupation,relationship,race,gender,capital_gain,capital_loss,hours_per_week,native_country]
```

```python
linear_classifier = tf.estimator.LinearClassifier(feature_columns=feat_cols,n_classes=2)
```

```python
train_input_func = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
linear_classifier.train(input_fn=train_input_func,steps=1000)
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=x_test,shuffle=False)
```

```python
pred_gen= linear_classifier.predict(input_fn=pred_input_func)
```

```python
prediction =list(pred_gen)
```

```python
prediction
```

```python
predictions = []
for i in prediction:
    predictions.append(i['class_ids'][0])
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
